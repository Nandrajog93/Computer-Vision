# Import Libraries

```python
import numpy as np
import cv2
from keras.preprocessing.image import img_to_array
from keras.models import load_model
from keras.utils import get_file
import argparse
import os
from matplotlib import pyplot as plt
```

# Pre-Trained Model

```python
dwnld_link = "https://s3.ap-south-1.amazonaws.com/arunponnusamy/pre-trained-weights/gender_detection.model"
model_path = get_file("gender_detection.model", dwnld_link, cache_subdir="pre-trained", cache_dir=os.getcwd())
classes= ['male', 'female']
```

# Classifier

```python
face_Classifier = cv2.CascadeClassifier('/Users/nandrajog/Downloads/opencv-master/data/haarcascades/haarcascade_frontalface_alt.xml')
eye_classifier = cv2.CascadeClassifier('/Users/nandrajog/haarcascade_eye.xml')
# smile_classifier = cv2.CascadeClassifier("/Users/nandrajog/Downloads/Face_eyes_smile-detection-master/smile.xml")
```

# Implementation

```python
model = load_model(model_path) #Gender detection

def face_detector(image):
    grey = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    face = face_Classifier.detectMultiScale(grey,1.3,5)
#     smile = smile_classifier.detectMultiScale(grey,1.2,5)
    eyes = eye_classifier.detectMultiScale(grey,1.3,5)
    
    
    for (x,y,w,h) in face:
        cv2.rectangle(image,(x,y),(x+w, y+h), (0,255,35),2)
        roi = image[y:y+h , x:x+w]
        print("detect faces", len(face))

        
        #Image Processing
        face_crop = cv2.resize(image, (96,96))
        face_crop = face_crop.astype("float")/255
        face_crop = img_to_array(face_crop)
        face_crop = np.expand_dims(face_crop,axis=0)


        store_accuracy=[]
        gender = model.predict(face_crop)[0]
        print(gender[0])
        accuracy = classes[np.argmax(gender)]
        print(accuracy)
        for i in gender:
            store_accuracy.append(gender)
        
#         print(accuracy)

        text = '%s '%(accuracy)
        cv2.putText(image,text,(x, y),cv2.FONT_HERSHEY_SIMPLEX , 1, (255, 255, 0), 2, cv2.LINE_AA)

        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(image,(ex,ey),(ex+ew,ey+eh),(0,0,255),3)
            print("detect eyes",len(eyes))
            
#     for (sx,sy,sw,sh) in smile:
#         cv2.rectangle(image,(sx,sy),(sx+sw,sy+sh),(0,255,255),3)
#         print("detect smile", len(smile))
#         print(store_accuracy)
        plt.plot(store_accuracy)
        plt.ylabel('Accuracy %')
        plt.show()
        
    return image



webcam = cv2.VideoCapture(0) 
while True:
    _,frame = webcam.read()
    
    
    cv2.imshow("frame",face_detector(frame))

    if cv2.waitKey(1)==13:break

        
webcam.release()
cv2.destroyAllWindows()
```
