```python
import cv2
import numpy as np 
```

```python
image = np.zeros((512,512,3) ,np.uint8)
```

```python
image = cv2.imread('/Users/nandrajog/alien/alien-07.jpg',0)
```

```python
image_1 = cv2.rectangle(image,(384,500),(510,128),(255,0,0),3)
image_1 = cv2.rectangle(image,(100,200),(510,128),(255,0,0),3)
image_1 = cv2.rectangle(image,(300,350),(510,128),(255,0,0),3)
```

```python
#Contours are uses for edge detection  so here we use sobel opertor  for edge detection 

canny = cv2.Canny(image_1,10,50)
```

```python
im2, contours, hierarchy= cv2.findContours(canny,cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cv2.imshow("image_1", image_1)
cv2.imshow("canny", canny)
cv2.imshow("contours", canny)

cv2.waitKey(0)
cv2.waitKey(0)
cv2.waitKey(0)

cv2.destroyAllWindows()
```
