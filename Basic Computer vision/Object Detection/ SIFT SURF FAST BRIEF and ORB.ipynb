In open cv 3.4.1 we can only use ORB 

```python
image = cv2.imread("/Users/nandrajog/07.jpg",0)
```

```python
## Initiate ORB detector
orb = cv2.ORB_create()
```

```python
# find the keypoints with ORB
keep_points = orb.detect(image,None)
```

```python
# compute the descriptors with ORB
keep_points, descriptors = orb.compute(image,keep_points)
```

```python
#creeateKeypoints

image_2 = cv2.drawKeypoints(image,keep_points,None,(124,0,255),0)
```

```python
cv2.imshow("image_2",image_2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
